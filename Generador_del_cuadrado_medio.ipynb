```python
from tabulate import tabulate

def generador_cuadrado_medio(semilla, iteraciones):
    """
    Implementa el generador del cuadrado medio

    Parámetros:
    - semilla: Número inicial (debe tener un número par de dígitos).
    - iteraciones: Número de iteraciones para generar números pseudoaleatorios.

    Retorna:
    - Una lista con las semillas, cuadrados (con ceros a la izquierda si fueron necesarios), números generados y números normalizados.
    """
    resultados = []
    longitud = len(str(semilla))  # Longitud de la semilla

    for i in range(iteraciones):
        cuadrado = semilla ** 2  # Elevar la semilla al cuadrado
        cuadrado_str = str(cuadrado).zfill(2 * longitud)  # Rellenar con ceros a la izquierda si es necesario
        medio = cuadrado_str[(len(cuadrado_str) - longitud) // 2 : (len(cuadrado_str) + longitud) // 2]  # Extraer el medio
        medio_int = int(medio)  # Convertir a entero
        normalizado = medio_int / (10 ** longitud)  # Normalizar el número entre 0 y 1

        # Guardar los resultados, incluyendo el cuadrado con ceros a la izquierda si fueron necesarios
        resultados.append([i + 1, semilla, cuadrado_str, medio, normalizado])

        semilla = medio_int  # Actualizar la semilla

        # Si la semilla es 0, el ciclo termina
        if semilla == 0:
            print(f"El ciclo terminó en la iteración {i + 1} porque la semilla es 0.")
            break

    return resultados

# Parámetros de entrada
semilla_inicial = 234566  # Semilla inicial (debe tener un número par de dígitos)
iteraciones = 100  # Número de iteraciones

# Generar los resultados
resultados = generador_cuadrado_medio(semilla_inicial, iteraciones)

# Imprimir la tabla usando tabulate
print(tabulate(resultados, headers=["Iteración", "Semilla", "Cuadrado", "Número Generado", "Normalizado"], tablefmt="pretty"))
```
