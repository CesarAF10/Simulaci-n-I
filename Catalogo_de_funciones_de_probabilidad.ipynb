## Distribución Uniforme

### Parámetros Generales
- **Intervalo**: $a$ (mínimo) y $b$ (máximo), donde $a < b$.
- **Notación estándar**: $X \sim U(a, b)$

### Funciones
1. **Función de Densidad de Probabilidad (PDF)**:
   $$
   f(x) = \begin{cases}
   \frac{1}{b - a} & \text{para } a \leq x \leq b \\
   0 & \text{en otro caso}
   \end{cases}
   $$

2. **Función de Distribución Acumulativa (CDF)**:
   $$
   F(x) = \begin{cases}
   0 & \text{para } x < a \\
   \frac{x - a}{b - a} & \text{para } a \leq x \leq b \\
   1 & \text{para } x > b
   \end{cases}
   $$

### Medidas Estadísticas
- **Media**: $\mu = \frac{a + b}{2}$
- **Varianza**: $\sigma^2 = \frac{(b - a)^2}{12}$
- **Desviación Estándar**: $\sigma = \frac{b - a}{\sqrt{12}}$

### Pseudocódigo para Simulación

    FUNCIÓN generar_uniforme(a, b):
    u ← ALEATORIO_ENTRE(0, 1)
    x ← a + u * (b - a)
    DEVOLVER x
    FIN FUNCIÓN

Código para simular la distribución:

    def simular_uniforme(a, b):
    u = np.random.random()  # Paso 1
    x = a + u * (b - a)     # Paso 2
    return x                # Paso 3

## Distribución Triangular

### Parámetros Generales
- **Límites**: $a$ (mínimo), $b$ (máximo), $c$ (moda), donde $a \leq c \leq b$.
- **Notación estándar**: $X \sim \text{Triangular}(a, c, b)$

### Funciones
1. **Función de Densidad de Probabilidad (PDF)**:
   $$
   f(x) = \begin{cases}
   \frac{2(x-a)}{(b-a)(c-a)} & \text{para } a \leq x \leq c \\
   \frac{2(b-x)}{(b-a)(b-c)} & \text{para } c < x \leq b \\
   0 & \text{en otro caso}
   \end{cases}
   $$

2. **Función de Distribución Acumulativa (CDF)**:
   $$
   F(x) = \begin{cases}
   0 & \text{para } x < a \\
   \frac{(x-a)^2}{(b-a)(c-a)} & \text{para } a \leq x \leq c \\
   1 - \frac{(b-x)^2}{(b-a)(b-c)} & \text{para } c < x \leq b \\
   1 & \text{para } x > b
   \end{cases}
   $$

### Medidas Estadísticas
- **Media**: $\mu = \frac{a + b + c}{3}$
- **Varianza**: $\sigma^2 = \frac{a^2 + b^2 + c^2 - ab - ac - bc}{18}$
- **Desviación Estándar**: $\sigma = \sqrt{\frac{a^2 + b^2 + c^2 - ab - ac - bc}{18}}$

### Pseudocódigo para Simulación

    FUNCIÓN generar_triangular(a, c, b):
    u ← ALEATORIO_ENTRE(0, 1)
    f_c ← 2/(b - a)  # Altura de la moda en PDF
    área_izq ← (c - a)/(b - a)
    SI u <= área_izq:
        x ← a + sqrt(u * (b - a) * (c - a))
    SI NO:
        x ← b - sqrt((1 - u) * (b - a) * (b - c))
    FIN SI
    DEVOLVER x
    FIN FUNCIÓN

**Código para simular la distribución**

    import numpy as np
    def simular_triangular(a, c, b):
    u = np.random.random()
    if u <= (c - a)/(b - a):
        return a + np.sqrt(u * (b - a) * (c - a))
    else:
        return b - np.sqrt((1 - u) * (b - a) * (b - c))
    Ejemplo de uso
    a, c, b = 1, 3, 5
    muestra = simular_triangular(a, c, b)

## Distribución Exponencial
### Parámetros Generales
- **Tasa de decaimiento**: $\lambda > 0$ (también llamado parámetro de tasa)
- **Notación estándar**: $X \sim \text{Exp}(\lambda)$

### Funciones
1. **Función de Densidad de Probabilidad (PDF)**:
   $$
   f(x) = \begin{cases}
   \lambda e^{-\lambda x} & \text{para } x \geq 0 \\
   0 & \text{para } x < 0
   \end{cases}
   $$

2. **Función de Distribución Acumulativa (CDF)**:
   $$
   F(x) = \begin{cases}
   1 - e^{-\lambda x} & \text{para } x \geq 0 \\
   0 & \text{para } x < 0
   \end{cases}
   $$

### Medidas Estadísticas
- **Media**: $\mu = \frac{1}{\lambda}$
- **Varianza**: $\sigma^2 = \frac{1}{\lambda^2}$
- **Desviación Estándar**: $\sigma = \frac{1}{\lambda}$

### Pseudocódigo para Simulación

    FUNCIÓN generar_exponencial(lambda):
    u ← ALEATORIO_ENTRE(0, 1)
    x ← -ln(1 - u)/lambda
    DEVOLVER x
    FIN FUNCIÓN

## Código para simulación

    import numpy as np
    def simular_exponencial(lambd):
    u = np.random.random()
    return -np.log(1 - u)/lambd
    # Ejemplo de uso
    lambda_param = 0.5
    muestra = simular_exponencial(lambda_param)

## Distribución m-Erlang

### Parámetros Generales
- **Forma (m)**: $m \in \mathbb{Z}^+$ (número de etapas)
- **Tasa ($\lambda$)**: $\lambda > 0$ (tasa por etapa)
- **Notación estándar**: $X \sim \text{Erlang}(m, \lambda)$

### Funciones
1. **Función de Densidad de Probabilidad (PDF)**:
   $$
   f(x) = \begin{cases}
   \frac{\lambda^m x^{m-1} e^{-\lambda x}}{(m-1)!} & \text{para } x \geq 0 \\
   0 & \text{para } x < 0
   \end{cases}
   $$

2. **Función de Distribución Acumulativa (CDF)**:
   $$
   F(x) = \begin{cases}
   1 - e^{-\lambda x} \sum_{k=0}^{m-1} \frac{(\lambda x)^k}{k!} & \text{para } x \geq 0 \\
   0 & \text{para } x < 0
   \end{cases}
   $$

### Medidas Estadísticas
- **Media**: $\mu = \frac{m}{\lambda}$
- **Varianza**: $\sigma^2 = \frac{m}{\lambda^2}$
- **Desviación Estándar**: $\sigma = \frac{\sqrt{m}}{\lambda}$

### Pseudocódigo para Simulación

    FUNCIÓN generar_erlang(m, lambda):
    x ← 0
    PARA i DESDE 1 HASTA m:
        u ← ALEATORIO_ENTRE(0, 1)
        x ← x - ln(1 - u)/lambda
    FIN PARA
    DEVOLVER x
    FIN FUNCIÓN

## Código para simulación
    import numpy as np
    def simular_erlang(m, lambd):
    return sum(-np.log(1 - np.random.random(m))/lambd
    # Ejemplo de uso
    m, lambda_param = 3, 0.5
    muestra = simular_erlang(m, lambda_param)

## Distribución Gamma

### Parámetros Generales
- **Forma (α)**: $α > 0$ (parámetro de forma)
- **Escala (β)**: $β > 0$ (parámetro de escala, donde $β = 1/λ$)
- **Notación estándar**: $X \sim \text{Gamma}(α, β)$

### Funciones
1. **Función de Densidad de Probabilidad (PDF)**:
   $$
   f(x) = \begin{cases}
   \frac{x^{α-1} e^{-x/β}}{β^α Γ(α)} & \text{para } x > 0 \\
   0 & \text{para } x \leq 0
   \end{cases}
   $$
   Donde $Γ(α)$ es la función gamma.

2. **Función de Distribución Acumulativa (CDF)**:
   $$
   F(x) = \frac{γ(α, x/β)}{Γ(α)}
   $$
   Donde $γ(α, x/β)$ es la función gamma incompleta.

### Medidas Estadísticas
- **Media**: $\mu = αβ$
- **Varianza**: $\sigma^2 = αβ^2$
- **Desviación Estándar**: $\sigma = β\sqrt{α}$

### Pseudocódigo para Simulación (Método de Ahrens-Dieter)

    FUNCIÓN generar_gamma(α, β):
    SI α < 1:
        // Usar método de transformación inversa modificado
        b ← (α + e)/e
        REPETIR:
            u ← ALEATORIO_ENTRE(0, 1)
            p ← b * u
            SI p <= 1:
                x ← p^(1/α)
                u1 ← ALEATORIO_ENTRE(0, 1)
                SI u1 <= e^(-x):
                    DEVOLVER x * β
            SI NO:
                x ← -ln((b - p)/α)
                u1 ← ALEATORIO_ENTRE(0, 1)
                SI u1 <= x^(α-1):
                    DEVOLVER x * β
            FIN SI
        FIN REPETIR
    SI NO:
        // Usar método de rechazo con distribución normal
        d ← α - 1/3
        c ← 1/sqrt(9*d)
        REPETIR:
            z ← ALEATORIO_NORMAL(0, 1)
            v ← (1 + c*z)^3
            SI v <= 0:
                CONTINUAR
            u ← ALEATORIO_ENTRE(0, 1)
            SI ln(u) <= 0.5*z^2 + d - d*v + d*ln(v):
                DEVOLVER d*v * β
        FIN REPETIR
    FIN SI
    FIN FUNCIÓN

## Código para simulación
    import numpy as np
    from math import e, sqrt, log

    def simular_gamma(alpha, beta):
    if alpha < 1:
        b = (alpha + e)/e
        while True:
            u = np.random.random()
            p = b * u
            if p <= 1:
                x = p ** (1/alpha)
                u1 = np.random.random()
                if u1 <= np.exp(-x):
                    return x * beta
            else:
                x = -log((b - p)/alpha)
                u1 = np.random.random()
                if u1 <= x ** (alpha - 1):
                    return x * beta
    else:
        d = alpha - 1/3
        c = 1/sqrt(9*d)
        while True:
            z = np.random.normal()
            v = (1 + c*z)**3
            if v <= 0:
                continue
            u = np.random.random()
            if log(u) <= 0.5*z**2 + d - d*v + d*log(v):
                return d*v * beta

    # Ejemplo de uso
    alpha, beta = 2.5, 1.0
    muestra = simular_gamma(alpha, beta)

## Distribución Weibull

### Parámetros Generales
- **Forma (k)**: $k > 0$ (parámetro de forma)
- **Escala (λ)**: $\lambda > 0$ (parámetro de escala)
- **Notación estándar**: $X \sim \text{Weibull}(k, \lambda)$

### Funciones
1. **Función de Densidad de Probabilidad (PDF)**:
   $$
   f(x) = \begin{cases}
   \frac{k}{\lambda} \left(\frac{x}{\lambda}\right)^{k-1} e^{-(x/\lambda)^k} & \text{para } x \geq 0 \\
   0 & \text{para } x < 0
   \end{cases}
   $$

2. **Función de Distribución Acumulativa (CDF)**:
   $$
   F(x) = \begin{cases}
   1 - e^{-(x/\lambda)^k} & \text{para } x \geq 0 \\
   0 & \text{para } x < 0
   \end{cases}
   $$

### Medidas Estadísticas
- **Media**: $\mu = \lambda \Gamma\left(1 + \frac{1}{k}\right)$
- **Varianza**: $\sigma^2 = \lambda^2 \left[\Gamma\left(1 + \frac{2}{k}\right) - \left(\Gamma\left(1 + \frac{1}{k}\right)\right)^2\right]$
- **Desviación Estándar**: $\sigma = \lambda \sqrt{\Gamma\left(1 + \frac{2}{k}\right) - \left(\Gamma\left(1 + \frac{1}{k}\right)\right)^2}$

### Pseudocódigo para Simulación
    FUNCIÓN generar_weibull(k, λ):
       u ← ALEATORIO_ENTRE(0, 1)
       x ← λ * (-ln(1 - u))^(1/k)
       DEVOLVER x
    FIN FUNCIÓN

## Código para simulación
    import numpy as np

    def simular_weibull(k, lambd):
      u = np.random.random()
      return lambd * (-np.log(1 - u))**(1/k)

    # Ejemplo de uso
    k, lambd = 1.5, 2.0
    muestra = simular_weibull(k, lambd)

## Distribución Normal

### Parámetros Generales
- **Media (μ)**: $-\infty < \mu < \infty$ (parámetro de ubicación)
- **Desviación estándar (σ)**: $\sigma > 0$ (parámetro de escala)
- **Notación estándar**: $X \sim N(\mu, \sigma^2)$

### Funciones
1. **Función de Densidad de Probabilidad (PDF)**:
   $$
   f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}
   $$

2. **Función de Distribución Acumulativa (CDF)**:
   $$
   F(x) = \frac{1}{2}\left[1 + \text{erf}\left(\frac{x-\mu}{\sigma\sqrt{2}}\right)\right]
   $$
   Donde erf es la función error.

### Medidas Estadísticas
- **Media**: $\mu$
- **Varianza**: $\sigma^2$
- **Desviación Estándar**: $\sigma$

### Pseudocódigo para Simulación (Método Box-Muller)

    FUNCIÓN generar_normal(μ, σ):
       REPETIR:
        u1 ← ALEATORIO_ENTRE(0, 1)
        u2 ← ALEATORIO_ENTRE(0, 1)
        z0 ← sqrt(-2*ln(u1)) * cos(2*π*u2)
        z1 ← sqrt(-2*ln(u1)) * sin(2*π*u2)
        x0 ← μ + σ*z0
        x1 ← μ + σ*z1
        DEVOLVER x0 (y opcionalmente x1 para otra muestra)
       FIN REPETIR
    FIN FUNCIÓN

### Código para simulación
    import numpy as np
    from math import sqrt, log, cos, sin, pi

    def simular_normal(mu, sigma):
      u1 = np.random.random()
      u2 = np.random.random()
      z0 = sqrt(-2*log(u1)) * cos(2*pi*u2)
      return mu + sigma*z0

    # Ejemplo de uso
    mu, sigma = 0, 1  # Distribución normal estándar
    muestra = simular_normal(mu, sigma)

## Distribución Log-Normal

### Parámetros Generales
- **Media logarítmica (μ)**: $-\infty < \mu < \infty$ (media del logaritmo natural de la variable)
- **Desviación estándar logarítmica (σ)**: $\sigma > 0$ (desviación estándar del logaritmo natural)
- **Notación estándar**: $X \sim \text{LogN}(\mu, \sigma^2)$

### Relación con la Normal
Si $Y \sim N(\mu, \sigma^2)$, entonces $X = e^Y \sim \text{LogN}(\mu, \sigma^2)$

### Funciones
1. **Función de Densidad de Probabilidad (PDF)**:
   $$
   f(x) = \begin{cases}
   \frac{1}{x\sigma\sqrt{2\pi}} e^{-\frac{(\ln x - \mu)^2}{2\sigma^2}} & \text{para } x > 0 \\
   0 & \text{para } x \leq 0
   \end{cases}
   $$

2. **Función de Distribución Acumulativa (CDF)**:
   $$
   F(x) = \Phi\left(\frac{\ln x - \mu}{\sigma}\right)
   $$
   Donde $\Phi$ es la CDF de la distribución normal estándar.

### Medidas Estadísticas
- **Media**: $e^{\mu + \sigma^2/2}$
- **Varianza**: $(e^{\sigma^2} - 1)e^{2\mu + \sigma^2}$
- **Moda**: $e^{\mu - \sigma^2}$

### Pseudocódigo para Simulación

    FUNCIÓN generar_lognormal(μ, σ):
      z ← ALEATORIO_NORMAL(0, 1)  # Generar normal estándar
      y ← μ + σ*z                 # Transformar a N(μ, σ²)
      x ← e^y                     # Aplicar exponencial
      DEVOLVER x
    FIN FUNCIÓN

### Código para simulación
    import numpy as np

    def simular_lognormal(mu, sigma):
      z = np.random.normal(0, 1)
      return np.exp(mu + sigma*z)

    # Versión vectorizada para múltiples muestras:
    def simular_lognormal(mu, sigma, size=1):
      return np.exp(mu + sigma*np.random.normal(0, 1, size))

    # Ejemplo de uso
    mu, sigma = 0, 1  # Parámetros del logaritmo
    muestra = simular_lognormal(mu, sigma)

## Distribución Beta

### Parámetros Generales
- **Forma (α)**: $\alpha > 0$ (primer parámetro de forma)
- **Forma (β)**: $\beta > 0$ (segundo parámetro de forma)
- **Notación estándar**: $X \sim \text{Beta}(\alpha, \beta)$

### Funciones
1. **Función de Densidad de Probabilidad (PDF)**:
   $$
   f(x) = \begin{cases}
   \frac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha, \beta)} & \text{para } 0 \leq x \leq 1 \\
   0 & \text{en otro caso}
   \end{cases}
   $$
   Donde $B(\alpha, \beta)$ es la función beta:
   $$
   B(\alpha, \beta) = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha + \beta)}
   $$

2. **Función de Distribución Acumulativa (CDF)**:
   $$
   F(x) = \frac{B(x; \alpha, \beta)}{B(\alpha, \beta)}
   $$
   Donde $B(x; \alpha, \beta)$ es la función beta incompleta.

### Medidas Estadísticas
- **Media**: $\mu = \frac{\alpha}{\alpha + \beta}$
- **Varianza**: $\sigma^2 = \frac{\alpha\beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}$
- **Moda**: $\text{moda} = \frac{\alpha - 1}{\alpha + \beta - 2}$ (para $\alpha, \beta > 1$)

### Pseudocódigo para Simulación (Método de Johnk)

    FUNCIÓN generar_beta(α, β):
         SI α < 1 Y β < 1:
           REPETIR:
            u ← ALEATORIO_ENTRE(0, 1)
            v ← ALEATORIO_ENTRE(0, 1)
            w ← u^(1/α)
            z ← v^(1/β)
            SI w + z <= 1:
                DEVOLVER w/(w + z)
           FIN REPETIR
         SI NO:
           // Usar método de transformación de Gamma
           g1 ← generar_gamma(α, 1)
           g2 ← generar_gamma(β, 1)
           DEVOLVER g1/(g1 + g2)
         FIN SI
    FIN FUNCIÓN

### Código para simulación
    import numpy as np

    def simular_beta(alpha, beta):
      if alpha < 1 and beta < 1:
        while True:
            u = np.random.random()
            v = np.random.random()
            w = u ** (1/alpha)
            z = v ** (1/beta)
            if w + z <= 1:
                return w / (w + z)
      else:
        g1 = np.random.gamma(alpha, 1)
        g2 = np.random.gamma(beta, 1)
        return g1 / (g1 + g2)

    # Ejemplo de uso
    alpha, beta = 2.5, 3.1
    muestra = simular_beta(alpha, beta)

## Distribuciones Pearson Tipos

### Clasificación General
El sistema de Pearson organiza distribuciones continuas en 7 tipos (I-VII) basados en sus momentos.

---

### Tipo I: Distribución Beta
- **Forma**: $f(x) \propto (x-a)^{p-1}(b-x)^{q-1}$
- **Parámetros**: $a, b$ (límites), $p, q > 0$ (formas)
- **Casos especiales**: Uniforme (p=q=1), Arcoseno (p=q=0.5)

---

### Tipo III: Gamma
- **Forma**: $f(x) \propto x^{k-1}e^{-x/\theta}$
- **Relación**: $\text{Gamma}(k, \theta) = \text{Pearson III}(k, \theta)$
- **Caso especial**: Cuando $k=1$ → Exponencial

---

### Tipo IV: Forma general asimétrica
- **Forma compleja**:
  $$
  f(x) \propto \left[1 + \left(\frac{x-\lambda}{a}\right)^2\right]^{-m} e^{-\nu \arctan\left(\frac{x-\lambda}{a}\right)}
  $$
- **Parámetros**: $m > 1/2$, $\nu$, $a > 0$, $\lambda$

---

### Tipo V: Inversa Gamma
- **Relación**: Si $X \sim \text{Gamma}(\alpha, \beta)$ entonces $1/X \sim \text{Pearson V}(\alpha, \beta)$
- **PDF**:
  $$
  f(x) = \frac{\beta^\alpha}{\Gamma(\alpha)} x^{-\alpha-1} e^{-\beta/x}
  $$

---

### Tipo VI: Beta Prime (F de Fisher)
- **Relación**: $\text{BetaPrime}(a,b) = \frac{X_1}{X_2}$ donde $X_1 \sim \text{Gamma}(a,1)$, $X_2 \sim \text{Gamma}(b,1)$
- **PDF**:
  $$
  f(x) = \frac{x^{a-1}(1+x)^{-a-b}}{B(a,b)}
  $$

---

### Tipo VII: t de Student
- **Forma**:
  $$
  f(x) \propto \left(1 + \frac{x^2}{\nu}\right)^{-(\nu+1)/2}
  $$
- **Parámetro**: $\nu > 0$ (grados de libertad)

---

### Pseudocódigo para Simulación (Tipo IV)

    FUNCIÓN generar_pearson_IV(m, ν, a, λ):
      REPETIR:
        y ← generar_normal(0, 1)
        u ← ALEATORIO_ENTRE(0, 1)
        densidad ← [1 + (y/a)^2]^(-m) * exp(-ν*arctan(y/a))
        SI u <= densidad / MAX_DENSIDAD:
            DEVOLVER λ + a*y
      FIN REPETIR
    FIN FUNCIÓN

### Código para simular (Tipo VII)

    import numpy as np

    def simular_pearson_VII(nu):
       z = np.random.normal(0, 1)
       v = np.random.chisquare(nu)
       return z / np.sqrt(v/nu)

    # Ejemplo uso
    nu = 5  # grados libertad
    muestra = simular_pearson_VII(nu)

## Distribución Log-Logística

### Parámetros Generales
- **Escala (α)**: $\alpha > 0$ (parámetro de escala)
- **Forma (β)**: $\beta > 0$ (parámetro de forma)
- **Notación estándar**: $X \sim \text{LL}(\alpha, \beta)$

### Relación con la Logística
Si $Y \sim \text{Logística}(\mu, \sigma)$, entonces $X = e^Y \sim \text{LL}(\alpha=e^\mu, \beta=\pi/\sigma\sqrt{3})$

### Funciones
1. **Función de Densidad de Probabilidad (PDF)**:
   $$
   f(x) = \begin{cases}
   \frac{(\beta/\alpha)(x/\alpha)^{\beta-1}}{\left[1 + (x/\alpha)^\beta\right]^2} & \text{para } x \geq 0 \\
   0 & \text{para } x < 0
   \end{cases}
   $$

2. **Función de Distribución Acumulativa (CDF)**:
   $$
   F(x) = \begin{cases}
   \frac{1}{1 + (x/\alpha)^{-\beta}} = \frac{(x/\alpha)^\beta}{1 + (x/\alpha)^\beta} & \text{para } x \geq 0 \\
   0 & \text{para } x < 0
   \end{cases}
   $$

### Medidas Estadísticas
- **Media**:
  $$
  \mu = \begin{cases}
  \frac{\alpha\pi/\beta}{\sin(\pi/\beta)} & \text{para } \beta > 1 \\
  \text{indefinida} & \text{para } \beta \leq 1
  \end{cases}
  $$
  
- **Varianza**:
  $$
  \sigma^2 = \begin{cases}
  \alpha^2\left[\frac{2\pi/\beta}{\sin(2\pi/\beta)} - \frac{(\pi/\beta)^2}{\sin^2(\pi/\beta)}\right] & \text{para } \beta > 2 \\
  \text{indefinida} & \text{para } \beta \leq 2
  \end{cases}
  $$

- **Mediana**: $\alpha$
- **Moda**: $\alpha\left(\frac{\beta-1}{\beta+1}\right)^{1/\beta}$ (para $\beta \geq 1$)

### Pseudocódigo para Simulación

    FUNCIÓN generar_loglogistica(α, β):
       u ← ALEATORIO_ENTRE(0, 1)
       x ← α * (u/(1 - u))^(1/β)
       DEVOLVER x
    FIN FUNCIÓN

### Código para simulación

    import numpy as np

    def simular_loglogistica(alpha, beta):
       u = np.random.random()
       return alpha * (u / (1 - u)) ** (1/beta)

    # Ejemplo de uso
    alpha, beta = 1.5, 2.0
    muestra = simular_loglogistica(alpha, beta)

## Sistema de Distribuciones Johnson

### Clasificación General
El sistema Johnson permite modelar distribuciones arbitrarias mediante transformaciones de la normal. Se compone de tres familias principales:

---

### 1. $S_L$ (Log-normal)
- **Transformación**: $Z = \gamma + \delta \ln(X - \xi)$
- **Dominio**: $X > \xi$
- **PDF**:
  $$
  f(x) = \frac{\delta}{(x-\xi)\sqrt{2\pi}} e^{-\frac{1}{2}[\gamma + \delta \ln(x-\xi)]^2}
  $$

---

### 2. $S_B$ (Acotada)
- **Transformación**: $Z = \gamma + \delta \ln\left(\frac{X - \xi}{\xi + \lambda - X}\right)$
- **Dominio**: $\xi < X < \xi + \lambda$
- **PDF**:
  $$
  f(x) = \frac{\delta \lambda}{2\pi (x-\xi)(\xi+\lambda-x)} e^{-\frac{1}{2}\left[\gamma + \delta \ln\left(\frac{x-\xi}{\xi+\lambda-x}\right)\right]^2}
  $$

---

### 3. $S_U$ (No acotada)
- **Transformación**: $Z = \gamma + \delta \sinh^{-1}\left(\frac{X - \xi}{\lambda}\right)$
- **Dominio**: $-\infty < X < \infty$
- **PDF**:
  $$
  f(x) = \frac{\delta}{\lambda\sqrt{2\pi}} \frac{1}{\sqrt{1 + \left(\frac{x-\xi}{\lambda}\right)^2}} e^{-\frac{1}{2}\left[\gamma + \delta \sinh^{-1}\left(\frac{x-\xi}{\lambda}\right)\right]^2}
  $$

---

### Parámetros Comunes
- $\xi$: Parámetro de ubicación
- $\lambda$: Parámetro de escala ($\lambda > 0$)
- $\gamma$: Parámetro de forma 1
- $\delta$: Parámetro de forma 2 ($\delta > 0$)

### Pseudocódigo para Simulación ($S_U$)

    FUNCIÓN generar_johnson_SU(γ, δ, ξ, λ):
       z ← ALEATORIO_NORMAL(0, 1)
       y ← γ + δ * sinh⁻¹((z - ξ)/λ)
       x ← ξ + λ * sinh((y - γ)/δ)
       DEVOLVER x
    FIN FUNCIÓN

### Código para simulación ($S_U$)

    import numpy as np
    from scipy.special import sinh, arcsinh

    def simular_johnson_SU(gamma, delta, xi, lamda):
       z = np.random.normal(0, 1)
       y = gamma + delta * arcsinh(z)
       x = xi + lamda * sinh(y/delta)
      return x

    # Ejemplo de uso:
    gamma, delta = 1.5, 2.0  # Parámetros de forma
    xi, lamda = 0.0, 1.0     # Parámetros de ubicación y escala
    muestra = simular_johnson_SU(gamma, delta, xi, lamda)

## Distribución Uniforme

### Parámetros Generales
- **Valor mínimo (a)**: $a \in \mathbb{Z}$
- **Valor máximo (b)**: $b \in \mathbb{Z}$ (con $b \geq a$)
- **Notación estándar**: $X \sim \text{Unif}\{a, a+1, \ldots, b\}$

### Función de Masa de Probabilidad (PMF)
$$
P(X = k) = \begin{cases}
\frac{1}{b - a + 1} & \text{para } k = a, a+1, \ldots, b \\
0 & \text{en otro caso}
\end{cases}
$$

### Función de Distribución Acumulativa (CDF)
$$
F(x) = \begin{cases}
0 & \text{para } x < a \\
\frac{\lfloor x \rfloor - a + 1}{b - a + 1} & \text{para } a \leq x \leq b \\
1 & \text{para } x > b
\end{cases}
$$

### Medidas Estadísticas
- **Media**: $\mu = \frac{a + b}{2}$
- **Varianza**: $\sigma^2 = \frac{(b - a + 1)^2 - 1}{12}$
- **Entropía**: $H(X) = \ln(b - a + 1)$

### Pseudocódigo para Simulación

    FUNCIÓN generar_uniforme_discreta(a, b):
       u ← ALEATORIO_ENTRE(0, 1)
       k ← a + ⌊u * (b - a + 1)⌋
       DEVOLVER k
    FIN FUNCIÓN

### Código para simulación

    import numpy as np

    def simular_uniforme_discreta(a, b):
       return np.random.randint(a, b + 1)  # b+1 porque randint es exclusivo en el límite superior

    # Ejemplo de uso:
    a, b = 1, 6  # Simula un dado justo
    muestra = simular_uniforme_discreta(a, b)

## Distribución Bernoulli

### Parámetros Generales
- **Probabilidad de éxito (p)**: $0 \leq p \leq 1$
- **Notación estándar**: $X \sim \text{Bernoulli}(p)$

### Función de Masa de Probabilidad (PMF)
$$
P(X = k) = \begin{cases}
p & \text{para } k = 1 \\
1-p & \text{para } k = 0 \\
0 & \text{en otro caso}
\end{cases}
$$

### Función de Distribución Acumulativa (CDF)
$$
F(x) = \begin{cases}
0 & \text{para } x < 0 \\
1-p & \text{para } 0 \leq x < 1 \\
1 & \text{para } x \geq 1
\end{cases}
$$

### Medidas Estadísticas
- **Media**: $\mu = p$
- **Varianza**: $\sigma^2 = p(1-p)$
- **Asimetría**: $\gamma_1 = \frac{1-2p}{\sqrt{p(1-p)}}$
- **Curtosis**: $\gamma_2 = \frac{1-6p(1-p)}{p(1-p)}$

### Pseudocódigo para Simulación

    FUNCIÓN generar_bernoulli(p):
       u ← ALEATORIO_ENTRE(0, 1)
       SI u <= p:
           DEVOLVER 1
       SINO:
           DEVOLVER 0
       FIN SI
    FIN FUNCIÓN

### Código para simulación
    import numpy as np

    def simular_bernoulli(p):
       return 1 if np.random.random() <= p else 0

    # Versión vectorizada para múltiples muestras:
    def simular_bernoulli(p, size=1):
       return (np.random.random(size) <= p).astype(int)

    # Ejemplo de uso:
    p = 0.3  # Probabilidad de éxito
    muestra = simular_bernoulli(p)



## Distribución Geométrica

### Parámetros Generales
- **Probabilidad de éxito (p)**: $0 < p \leq 1$
- **Notación estándar**: $X \sim \text{Geom}(p)$

### Función de Masa de Probabilidad (PMF)
$$
P(X = k) = \begin{cases}
p(1-p)^{k-1} & \text{para } k = 1, 2, 3, \ldots \\
0 & \text{en otro caso}
\end{cases}
$$

### Función de Distribución Acumulativa (CDF)
$$
F(k) = \begin{cases}
0 & \text{para } k < 1 \\
1 - (1-p)^{\lfloor k \rfloor} & \text{para } k \geq 1
\end{cases}
$$

### Medidas Estadísticas
- **Media**: $\mu = \frac{1}{p}$
- **Varianza**: $\sigma^2 = \frac{1-p}{p^2}$
- **Asimetría**: $\gamma_1 = \frac{2-p}{\sqrt{1-p}}$
- **Curtosis**: $\gamma_2 = 6 + \frac{p^2}{1-p}$

### Pseudocódigo para Simulación

    FUNCIÓN generar_geometrica(p):
       k ← 1
       MIENTRAS ALEATORIO_ENTRE(0, 1) > p:
           k ← k + 1
       FIN MIENTRAS
       DEVOLVER k
    FIN FUNCIÓN

## Código para simulación

    import numpy as np

    def simular_geometrica(p):
       return np.random.geometric(p)

    # Ejemplo de uso:
    p = 0.2  # Probabilidad de éxito
    muestra = simular_geometrica(p)

## Distribución Binomial Negativa

### Parámetros Generales
- **Número de éxitos requeridos (r)**: $r \in \mathbb{Z}^+$
- **Probabilidad de éxito (p)**: $0 < p < 1$
- **Notación estándar**: $X \sim \text{NB}(r, p)$

### Función de Masa de Probabilidad (PMF)
$$
P(X = k) = \begin{cases}
\binom{k+r-1}{k} p^r (1-p)^k & \text{para } k = 0, 1, 2, \ldots \\
0 & \text{en otro caso}
\end{cases}
$$

### Función de Distribución Acumulativa (CDF)
$$
F(k) = \begin{cases}
0 & \text{para } k < 0 \\
I_p(r, \lfloor k \rfloor + 1) & \text{para } k \geq 0
\end{cases}
$$
donde $I_p$ es la función beta incompleta regularizada.

### Medidas Estadísticas
- **Media**: $\mu = \frac{r(1-p)}{p}$
- **Varianza**: $\sigma^2 = \frac{r(1-p)}{p^2}$
- **Asimetría**: $\gamma_1 = \frac{2-p}{\sqrt{r(1-p)}}$
- **Curtosis**: $\gamma_2 = \frac{p^2-6p+6}{r(1-p)}$

### Pseudocódigo para Simulación

    FUNCIÓN generar_binomial_negativa(r, p):
       éxitos ← 0
       intentos ← 0
       MIENTRAS éxitos < r:
           u ← ALEATORIO_ENTRE(0, 1)
           SI u <= p:
               éxitos ← éxitos + 1
           FIN SI
           intentos ← intentos + 1
       FIN MIENTRAS
       DEVOLVER intentos - r
    FIN FUNCIÓN

### Código para simulación

    import numpy as np

    def simular_binomial_negativa(r, p):
       return np.random.negative_binomial(r, p)

    # Ejemplo de uso:
    r, p = 5, 0.3  # 5 éxitos deseados con probabilidad 0.3
    muestra = simular_binomial_negativa(r, p)

## Distribución Poisson

### Parámetros Generales
- **Tasa de ocurrencia (λ)**: $\lambda > 0$ (parámetro de tasa)
- **Notación estándar**: $X \sim \text{Poisson}(\lambda)$

### Función de Masa de Probabilidad (PMF)
$$
P(X = k) = \begin{cases}
\frac{e^{-\lambda} \lambda^k}{k!} & \text{para } k = 0, 1, 2, \ldots \\
0 & \text{en otro caso}
\end{cases}
$$

### Función de Distribución Acumulativa (CDF)
$$
F(k) = e^{-\lambda} \sum_{i=0}^{\lfloor k \rfloor} \frac{\lambda^i}{i!}
$$

### Medidas Estadísticas
- **Media**: $\mu = \lambda$
- **Varianza**: $\sigma^2 = \lambda$
- **Asimetría**: $\gamma_1 = \lambda^{-1/2}$
- **Curtosis**: $\gamma_2 = \lambda^{-1}$

### Pseudocódigo para Simulación (Método de Knuth)

    FUNCIÓN generar_poisson(λ):
        L ← e^{-λ}
        p ← 1.0
        k ← 0
        REPETIR:
            k ← k + 1
            u ← ALEATORIO_ENTRE(0, 1)
            p ← p * u
        HASTA QUE p ≤ L
        DEVOLVER k - 1
    FIN FUNCIÓN

### Código para simulación

    import numpy as np
    from math import exp

    def simular_poisson(lambd):
        # Método directo para λ pequeña
        if lambd < 30:
           L = exp(-lambd)
           k = 0
           p = 1.0
           while p > L:
               k += 1
               p *= np.random.random()
           return k - 1
        else:
            # Aproximación normal para λ grande
            return int(np.random.normal(lambd, sqrt(lambd)))

    # Ejemplo de uso:
    lambd = 3.5  # Tasa de ocurrencia
    muestra = simular_poisson(lambd)
