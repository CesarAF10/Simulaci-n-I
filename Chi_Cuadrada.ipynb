```python
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Prueba Chi_Cuadrada

# Datos de la tabla
data = np.array([
    [0.048, 0.591, 0.478, 0.166, 0.932],
    [0.926, 0.007, 0.551, 0.362, 0.533],
    [0.368, 0.437, 0.178, 0.411, 0.545],
    [0.787, 0.594, 0.086, 0.951, 0.772],
    [0.298, 0.199, 0.157, 0.401, 0.631]
])

# Aplanar la matriz para obtener una lista unidimensional
data_flat = data.flatten()

# Definir el número de intervalos (bins) para la prueba Chi-Cuadrada
num_bins = 10

# Crear intervalos (bins) para clasificar los datos
observed_frequencies, bin_edges = np.histogram(data_flat, bins=num_bins, range=(0, 1))

# Frecuencias esperadas para una distribución uniforme
expected_frequency = len(data_flat) / num_bins
expected_frequencies = np.full(num_bins, expected_frequency)

# Realizar la prueba de Chi-Cuadrada
chi2_stat, p_value = chisquare(observed_frequencies, expected_frequencies)

# Mostrar los resultados
print("----- Prueba de Chi-Cuadrada -----")
print(f"Estadístico de Chi-Cuadrada: {chi2_stat}")
print(f"Valor p: {p_value}")

# Interpretación del resultado
if p_value > 0.05:
    print("No se rechaza la hipótesis nula: los datos siguen una distribución uniforme.")
else:
    print("Se rechaza la hipótesis nula: los datos no siguen una distribución uniforme.")

# Graficar frecuencias observadas y esperadas
plt.figure(figsize=(8, 6))
plt.bar(range(num_bins), observed_frequencies, alpha=0.7, label='Frecuencias observadas', color='blue')
plt.plot(range(num_bins), expected_frequencies, 'r--', label='Frecuencias esperadas', linewidth=2)
plt.title('Prueba de Chi-Cuadrada: Frecuencias observadas vs esperadas')
plt.xlabel('Intervalos')
plt.ylabel('Frecuencia')
plt.xticks(range(num_bins), [f"{bin_edges[i]:.1f}-{bin_edges[i+1]:.1f}" for i in range(num_bins)], rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
